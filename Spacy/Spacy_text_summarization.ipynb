```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
# Loading the English language model for spaCy
nlp = spacy.load('en_core_web_sm')

# Tokenizing the text into sentences using spaCy
doc = nlp(text_article4)
sentence_tokens = [sent for sent in doc.sents]
```

```python
# Calculating the selection length as 30% of the total number of sentences
select_length = int(len(sentence_tokens)*0.3)
```

```python
# Creating a dictionary of sentence scores based on word frequencies
word_frequencies = {}
for word in doc:
    if word.text.lower() not in STOP_WORDS and word.text.lower() not in punctuation:
        if word.text not in word_frequencies.keys():
            word_frequencies[word.text] = 1
        else:
            word_frequencies[word.text] += 1

max_frequency = max(word_frequencies.values())
for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/max_frequency)

sentence_scores = {}
for sent in sentence_tokens:
    for word in sent:
        if word.text.lower() in word_frequencies.keys():
            if sent not in sentence_scores.keys():
                sentence_scores[sent] = word_frequencies[word.text.lower()]
            else:
                sentence_scores[sent] += word_frequencies[word.text.lower()]
```

```python
# Generating a summary by selecting the most important sentences
from heapq import nlargest
summary = nlargest(select_length, sentence_scores, key = sentence_scores.get)
final_summary = [word.text for word in summary]
summary = ' '.join(final_summary)
```

```python
print("Original Text:\n", text_article4)
print("\nSummary:\n", summary)
```

### Here, I am setting key argument to a lambda function that takes a sentence from the sentence_scores dictionary and returns its length.
This function is used to determine the order of elements in the list returned by nlargest, so the resulting summary will consist of the longest sentences in the original text.

```python
# Generating a summary by selecting the most important sentences
from heapq import nlargest
summary = nlargest(select_length, sentence_scores, key = lambda x: len(x))
final_summary = [word.text for word in summary]
summary = ' '.join(final_summary)
```

```python
print("Original Text:\n", text_article4)
print("\nSummary:\n", summary)
```
